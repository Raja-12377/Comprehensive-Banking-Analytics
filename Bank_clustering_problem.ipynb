```python
# Standardize the data\

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Monthly_Balance']] = scaler.fit_transform(df[['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Monthly_Balance']])

# Round the scaled data to one digit
df[['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Monthly_Balance']] = df[['Age', 'Annual_Income', 'Monthly_Inhand_Salary', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Monthly_Balance']].applymap(lambda x: round(x, 0))
df=df.drop(['Payment_Behaviour'], axis=1)
df.tail()
```

```python
# Elbow Method to find the optimal number of clusters
from sklearn.cluster import KMeans
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 11), sse)
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Initialize a list to store the silhouette scores
silhouette_scores = []

# Iterate over the range of possible cluster numbers
for k in range(1, 11):
    # Perform K-Means clustering with the current number of clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)

    # Calculate the silhouette score for the current clustering
    if k > 1:
        silhouette_score_value = silhouette_score(df, kmeans.labels_)
    else:
        silhouette_score_value = None

    # Append the silhouette score to the list
    silhouette_scores.append(silhouette_score_value)

# Plot the silhouette scores
plt.plot(range(1, 11), silhouette_scores)
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.show()
```

```python
# Perform K-Means clustering with the best k cluster value
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(df)

# Assign cluster labels to the DataFrame
df['Cluster'] = kmeans.labels_

# Print the cluster labels
print(df['Cluster'])
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score for the clusters
silhouette_score = silhouette_score(df, df['Cluster'], metric='euclidean')

print("Silhouette Score:", silhouette_score)
```

```python
import seaborn as sns

# Create a scatter plot with different colors for each cluster using hue
sns.scatterplot(x='Age', y='Annual_Income', hue='Cluster', data=df)
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.title('K-Means Clustering Result')
plt.show()
```

```python

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Age'], df['Annual_Income'], df['Monthly_Inhand_Salary'], c=df['Cluster'])
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income')
ax.set_zlabel('Monthly Inhand Salary')
plt.title('K-Means Clustering Result')
plt.show()
```

```python
df=df.drop(['Cluster'], axis=1)
df.head()
```

```python
from sklearn.cluster import DBSCAN

# Perform DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(df)

# Get the cluster labels
labels = dbscan.labels_

# Assign cluster labels to the DataFrame
df['dbCluster'] = labels

# Print the cluster labels
print(df['dbCluster'])
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score for the clusters
silhouette_score = silhouette_score(df, df['dbCluster'], metric='euclidean')

print("Silhouette Score:", silhouette_score)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot with different colors for each cluster
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Annual_Income'], c=df['dbCluster'])
plt.xlabel('Age')
plt.ylabel('Annual Income')
plt.title('DBSCAN Clustering Result')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Perform agglomerative clustering
agglomerative = AgglomerativeClustering(n_clusters=4, linkage='ward')
agglomerative.fit(df)

# Assign cluster labels to the DataFrame
df['ACluster'] = agglomerative.labels_

# Print the cluster labels
print(df['ACluster'])
```

```python
from sklearn.metrics import silhouette_score

# Calculate the silhouette score for the clusters
silhouette_score = silhouette_score(df, df['ACluster'], metric='euclidean')

print("Silhouette Score:", silhouette_score)
```
