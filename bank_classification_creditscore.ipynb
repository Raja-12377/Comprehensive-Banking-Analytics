```python
#there is no missing values
#Feature selection and Target selection 
features = df[["Annual_Income", "Monthly_Inhand_Salary", 
                   "Num_Bank_Accounts", "Num_Credit_Card", 
                   "Interest_Rate", "Num_of_Loan", 
                   "Delay_from_due_date", "Num_of_Delayed_Payment", 
                   "Credit_Mix", "Outstanding_Debt", 
                   "Credit_History_Age", "Monthly_Balance"]]
target = df[["Credit_Score"]]
features.head()
```

```python
target.head()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder

# Separate numerical features and categorical features
numerical_features = features[["Annual_Income", "Monthly_Inhand_Salary", 
                              "Num_Bank_Accounts", "Num_Credit_Card", 
                              "Interest_Rate", "Num_of_Loan", 
                              "Delay_from_due_date", "Num_of_Delayed_Payment", 
                              "Outstanding_Debt", 
                              "Credit_History_Age", "Monthly_Balance"]]
categorical_features = features[["Credit_Mix"]]

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the numerical data and transform it
numerical_features_scaled = scaler.fit_transform(numerical_features)

# Convert the scaled data back to a DataFrame
numerical_features_scaled = pd.DataFrame(numerical_features_scaled, columns=numerical_features.columns)
numerical_features_scaled.head()
```

```python
# Create a LabelEncoder object
le = LabelEncoder()

# Fit the encoder to the categorical data and transform it
categorical_features_encoded = categorical_features.apply(le.fit_transform)

# Concatenate the numerical and categorical features
features = pd.concat([numerical_features_scaled, categorical_features_encoded], axis=1)

# Print the concatenated DataFrame
features.head()
```

```python
#feature scaling AND Encoding done 
#Our Target column is Multi class classification 
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(features, target, 
                                                    test_size=0.33, 
                                                    random_state=42)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()
dt_model.fit(xtrain, ytrain)
```

```python
dt_y_pred=dt_model.predict(xtest)
dt_y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report
# Calculate accuracy
accuracy = accuracy_score(ytest, dt_y_pred)
print(f"Accuracy: {accuracy:.2f}")

# Generate classification report
print(classification_report(ytest, dt_y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()
rf_model.fit(xtrain, ytrain)
```

```python
rf_y_pred=rf_model.predict(xtest)
rf_y_pred
```
